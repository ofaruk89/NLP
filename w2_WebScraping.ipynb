```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops'
request = requests.get(url)
soup = BeautifulSoup(request.text, "html.parser")

# Find all product containers
products = soup.find_all('div', class_='col-md-4 col-xl-4 col-lg-4')

for product in products:
    title = product.find('a', class_='title').get_text(strip=True)
    price = product.find('h4', class_='price').get_text(strip=True)
    description = product.find('p', class_='description').get_text(strip=True)
    review_count = product.find('p', class_='review-count').get_text(strip=True)
    data_rating = product.find('p', attrs={'data-rating': True})['data-rating']
    
    print(f'Title: {title}')
    print(f'Price: {price}')
    print(f'Description: {description}')
    print(f'Review Count: {review_count}')
    print(f'Rating: {data_rating} stars')
    print('-' * 40)
```

```python


# URL to scrape
url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops'
request = requests.get(url)
soup = BeautifulSoup(request.text, "html.parser")

#print(soup)

# Lists to hold the extracted data
titles = []
prices = []
descriptions = []
review_counts = []
ratings = []

# Find all product containers
products = soup.find_all('div', class_='col-md-4 col-xl-4 col-lg-4')

for product in products:
    title = product.find('a', class_='title').get_text(strip=True)
    price = product.find('h4', class_='price').get_text(strip=True)
    description = product.find('p', class_='description').get_text(strip=True)
    review_count = product.find('p', class_='review-count').get_text(strip=True)
    data_rating = product.find('p', attrs={'data-rating': True})['data-rating']
    
    # Append the extracted data to the lists
    titles.append(title)
    prices.append(price)
    descriptions.append(description)
    review_counts.append(review_count)
    ratings.append(data_rating)

# Create a DataFrame with specific columns
product_df = pd.DataFrame(zip(titles, prices, descriptions, review_counts, ratings), 
                          columns=['Titles', 'Prices', 'Descriptions', 'Review Counts', 'Ratings'])

# Show the first few rows of the DataFrame
print(product_df.head())

# Save to a CSV file
product_df.to_csv("ecommerce.csv", index=False)
```
